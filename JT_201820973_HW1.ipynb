```python
# append_b 함수 생성 (Ax = b 꼴 이라고 생각했을 때, A와 b를 붙힌 형태를 return)

def append_b(mat_A, mat_b):
    
    # row의 길이가 다르면 에러 출력
    
    assert len(mat_A) == len(mat_b), "함수의 길이가 맞지 않습니다."
    
    # mat_A의 row만큼 반복한다.
    
    for row in range(len(mat_A)):
        
        # 각 row의 mat_A에 mat_b의 수를 추가한다.
        
        mat_A[row].append(mat_b[row][0])
    
    # 만들어준 함수를 리턴한다.
    
    return mat_A
```

```python
# make_upper_triangle 함수 생성 (Ax = b 꼴 이라고 생각했을 때, A와 b를 붙힌 형태로 input)

def make_upper_triangle(mat):
    
    # 리턴할 리스트 생성
    
    ret_mat = [] 
    
    # 행렬의 행 수 저장 (연산이 완료된 행은 삭제하며 진행)

    row_index = list(range(len(mat))) 

    # 행렬의 열 수 저장 (마지막 col은 b값을 담고있는 줄 이므로 - 1)
    
    col_index = len(mat[0]) - 1 
    
    # 행렬의 열 수 만큼 반복한다.
    
    for col in range(col_index):
    
        # 0이 아닌 줄의 인덱스를 저장한다.
        
        rows_with_nonzero = [row for row in row_index if mat[row][col] != 0]
        
        """
        
        col열 중에 맞는 0이 아닌 row을 찾아 리턴할 리스트에 추가하고
        
        해당 col이 0이 아닌 나머지 row들과 연산하여 나머지 row들의 col을 0으로 만드는 과정
        
        """
        
        # row_with_nonzero가 비어있는 리스트가 아니라면
        
        if rows_with_nonzero:
            
            # row_with_nonzero의 첫 인덱스(원래 행렬의 행:mat[row][col]이 0 아닌 행)를 저장한다.
            
            pivot = rows_with_nonzero[0]
            
            # 리턴할 리스트에 방금 저장한 인덱스위치의 행을 추가한다.
            
            ret_mat.append(mat[pivot])
            
            # 방금 추가한 인덱스의 헹을 row_index에서 제거한다
            
            row_index.remove(pivot)
            
            # 제거하고 남은 col열 중에서 0이 아닌 행들동안 반복
            
            for row in rows_with_nonzero:
                
                # 만약 행이 방금 ret_mat에 추가한 행이 아니라면
                
                if row is not pivot:
                    
                    # multiplier : row행의 col을 0으로 만들어주기 위해 mat[pivot] 에 곱해줄 수
                    
                    multiplier = mat[row][col] / mat[pivot][col]
                    
                    # row행의 col을 0으로 만들어주는 작업
                    
                    mat[row] = [a - multiplier*b for a, b in zip(mat[row], mat[pivot])]           
    
    # 만약 위 과정에서 택해지지 않은 행이 있다면 (한 행이 모두 0인 row)
    
    for row in row_index:
            
        # ret_mat에 추가한다.
        
        ret_mat.append(mat[row])
    
    # ret_mat을 리턴한다.
    
    return ret_mat
```

```python
# ret_mat함수 생성 (을 return하는 함수)

def ret_mat(mat):
    
    # return할 리스트 생성
    
    ret_mat = []
    
    # 행렬의 행 수 저장 (연산이 완료된 행은 삭제하며 진행)
    
    row_index = list(range(len(mat)))
    
    # 행렬의 열 수 저장 (마지막 col은 b값을 담고있는 줄 이므로 - 1)
    
    col_index = len(mat[0]) - 1
    
    # 행렬의 열 수 만큼 반복한다. (뒤집어서 반복한다.)
    
    for col in range(col_index-1, 0, -1):
        
        # 0이 아닌 줄의 인덱스를 저장한다.
        
        rows_with_nonzero = [row for row in row_index if mat[row][col] != 0]
        
        """
        
        col열 중에 맞는 0이 아닌 row을 찾아 리턴할 리스트에 추가하고
        
        해당 col이 0이 아닌 나머지 row들과 연산하여 나머지 row들의 col을 0으로 만드는 과정
        
        위의 make_upper_triangle코드와 비슷하지만 이번 함수는 아래에서부터 제거하는 형태
        
        """
        
        # row_with_nonzero가 비어있는 리스트가 아니라면
        
        if rows_with_nonzero:
            
            # row_with_nonzero의 마지막 인덱스를 저장한다.
            
            pivot = rows_with_nonzero[-1]
            
            # 리턴할 리스트에 방금 저장한 인덱스위치의 행을 추가한다.
            
            ret_mat.append(mat[pivot])
            
            # 방금 추가한 인덱스의 헹을 row_index에서 제거한다
            
            row_index.remove(pivot)
            
            # 제거하고 남은 col열 중에서 0이 아닌 행들동안 반복
            
            for row in rows_with_nonzero:
                
                # multiplier : row행의 col을 0으로 만들어주기 위해 mat[pivot] 에 곱해줄 수
                
                multiplier = mat[row][col] / mat[pivot][col]
                
                 # row행의 col을 0으로 만들어주는 작업
                
                mat[row] = [a - multiplier*b for a, b in zip(mat[row], mat[pivot])]
        
    # 만약 위 과정에서 택해지지 않은 행이 있다면 (한 행이 모두 0인 row)  
    
    for row in row_index:
            
        # ret_mat에 추가한다.
        
        ret_mat.append(mat[row])
        
    # ret_mat을 뒤집어서 리턴한다. 
    
    return list(reversed(ret_mat))
```

```python
# Gauss_Naive 함수 생성

def Gauss_Naive(mat_A, mat_b):
    
    # return할 리스트 생성
    
    ret_x = []
    
    # A와 b를 합친 형태로 upper_triangle 모양으로 만들어줌
    
    upper_triangle = make_upper_triangle(append_b(mat_A, mat_b))
    
    # 역삼각행렬 모양의 행렬로 만들어줌
    
    mat = ret_mat(upper_triangle)
    
    # 행렬의 각 줄마다 반복
    
    for row in mat:
        
        # row인 값들중에 0이 아닌 값을 찾음
        
        for val in row[:-1]:
            
            # 0이 아닌 값을 찾으면
            
            if val != 0:
                
                # b 값에 나눠서 ret_x리스트에 추가해라
                
                ret_x.append([row[-1] / val])
                
    # ret_x를 리턴해라
                
    return ret_x
```

```python
mat_A = [[2, 1, 1],
        [4, -6, 0],
        [-2, 7, 2]]

mat_b = [[5],
         [-2],
         [9]]

print(Gauss_Naive(mat_A, mat_b))
```
